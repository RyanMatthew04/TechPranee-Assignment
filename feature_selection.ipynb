```python
import pandas as pd

df=pd.read_csv('manufacturing_defect_dataset.csv')
df.head(5)
```

```python
from ydata_profiling import ProfileReport
prof = ProfileReport(df)
prof.to_file(output_file='EDA.html')
```

```python
from sklearn.tree import DecisionTreeClassifier

X=df.drop(columns=['DefectStatus'])
y=df['DefectStatus']

model = DecisionTreeClassifier()
model.fit(X, y)

importance = model.feature_importances_

important_features = sorted(
    [(X.columns[i], importance[i]) for i in range(len(importance)) if importance[i] > 0.1],
    key=lambda x: x[1],
    reverse=True
)

for feature, imp in important_features:
    print(f"Feature: {feature}, Importance: {imp:.4f}")
```
